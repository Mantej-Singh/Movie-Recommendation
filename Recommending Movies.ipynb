## Importing "u data set" and "u.item set" Information about the movies

```python
#MovieLens data taken from
#https://grouplens.org/datasets/movielens/100k/
# Column names available in the readme file
```

```python
#Reading The full u data set, and selecting only 3 colums; till rating
cols = ['user_id', 'movie_id', 'rating']
data = pd.read_csv('C:/Users/Jarvis/MovieLens data/u.data', sep='\t', names=cols,encoding='latin-1', usecols=range(3))
```

```python
data.head(2)
```

```python
#Now, read iteam data set: Information about the items (movies); this is a tab separated list
cols = ['movie_id', 'movie_title' ,'release_date']
movies = pd.read_csv('C:/Users/Jarvis/MovieLens data/u.item', sep='|', names=cols,encoding='latin-1', usecols=range(3))
```

```python
movies.head(3)
```

```python
ratings = pd.merge(movies, data)
```

```python
ratings.head()
```

```python
moviepivots=ratings.pivot_table(index=['user_id'],columns=['movie_title'],values='rating', aggfunc=np.mean)
moviepivots.head()
```

```python
#I love Disney, so.....
disneyratings=moviepivots['Aladdin (1992)']
disneyratings.head()
```

# Building a correlation statistics score: Disney MOvies vs Others

```python
similardisneymovie=moviepivots.corrwith(disneyratings)
```

```python
similardisneymovie = similardisneymovie.dropna()
simi = pd.DataFrame(similardisneymovie)
simi.head(20)
```

```python
similardisneymovie.sort_values(ascending=False).head()
```

#### Lets improve our engine by increasing number of movies viewed 

```python
moviestats = ratings.groupby('movie_title').agg({'rating': [np.count_nonzero,np.sum,np.mean]})
moviestats.head()
```

#### Cross check our results 

```python
ratings[ratings.movie_title=='1-900 (1994)']
```

## selecting movies viewed my more number of people eg:150

```python
popularmovies = moviestats['rating']['count_nonzero'] >= 120
popularmovies.head(10)
```

```python
moviestats[popularmovies].sort_values([('rating', 'mean')], ascending=False).head()
```

```python
#moviestats['rating']['count_nonzero'] >= 150
```

```python
df = moviestats[popularmovies].join(pd.DataFrame(similardisneymovie, columns=['similarity']))
df.sort_values(['similarity'], ascending=False).head(20)
```
